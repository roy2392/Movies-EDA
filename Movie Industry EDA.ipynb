```python
# First let's import the packages we will use in this project
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
#Read in the data
df=pd.read_csv('movies.csv')
```

```python
#let's look at the data
df.head()
```

## Data Cleaning

```python
#check precents of missing data
for col in df.columns: 
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
# drop all rows with any NaN and NaT values
df = df.dropna()
print(df)
```

```python
# what are the data types of the columns
df.dtypes
```

```python
#change columns data types 
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
df
```

```python
#create a column with the correct year of each movie 
df['yearcorrect'] = df['released'].astype(str).str[:4]
```

```python
#in order to understand the data, we will want to absorve it by the gross
df.sort_values(by=['gross'] , inplace = True , ascending = False)
df
```

```python
#drop duplicates
df.drop_duplicates()
```

# Univariate Analysis

### 1. Hypotesis - Gross & Budegt are correlated

```python
#plot budget vs gross using seaborn
sns.regplot(x='budget', y= 'gross' , data = df,scatter_kws={"color" : "red"}
,line_kws = {"color":"blue"})
plt.title ('Gross vs Budget')
```

```python
#what is th excat correlation?
df.corr(method='pearson')
```

<b><div style ='background:lightgreen;dont-size:15px'> It seems that the gross & budget correlation is the highest</div></b>

```python
#create correlation matrix to all features
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot = True)
plt.title ('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

### 2. Hypotesis - Company & gross are correlated

```python
#convert objct datatypes to category 
df_1 = df
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
      df_numerized[col_name]=df_numerized[col_name].astype('category')
      df_numerized[col_name]=df_numerized[col_name].cat.codes
        
df_numerized     
```

```python
#create correlation matrix to all features
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot = True)
plt.title ('Correlation Matrix for All Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# check the correlation 
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
#sort asc
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#show correlations which are higher than 0.5
high_correlation = sorted_pairs[(sorted_pairs) > 0.5]
high_correlation
                    
```

<b><div style ='background:lightgreen;dont-size:15px'> In conclusion: gross & votes has the highest correlations. <br>
